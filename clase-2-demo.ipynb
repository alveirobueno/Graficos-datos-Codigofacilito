## Barras+

```python
def crear_grafico(datos, pais = 'Mexico'):
    subset = datos[datos['country'] == pais]
    barras_pais = alt.Chart(subset).mark_bar().encode(
            x = alt.X('year:O', title = '', axis = alt.Axis(labelAngle=0)),
            y = alt.Y('life_expect', title = '')
        ).properties(
            width = 400, 
            title = f"Esperanza de vida en {pais}"
        )

    return barras_pais
```

```python
crear_grafico(datos = data, pais = 'Dominican Republic')
```

```python
data = vega_data.gapminder_health_income()
data
```

```python
latinoamerica_countries = ["Bolivia","Argentina","Brazil","Chile","Colombia","Costa Rica","Cuba","Dominican Republic","Ecuador","El Salvador","French Guiana","Guadeloupe","Guatemala","Haiti","Honduras","Martinique","Mexico","Nicaragua","Panama","Paraguay","Peru","Puerto Rico","Saint Barthélemy","Saint Martin","Uruguay","Venezuela"]
top_10_latinoamerica_by_income = data[data['country'].isin(latinoamerica_countries)].sort_values(by='income', ascending=False)[:10]
top_10_latinoamerica_by_income['income'] = top_10_latinoamerica_by_income['income'].apply(lambda x: round(x/1_000, 1))
top_10_latinoamerica_by_income['population'] = top_10_latinoamerica_by_income['population'].apply(lambda x: round(x/1_000_000, 1))
print(top_10_latinoamerica_by_income)
alt.Chart(top_10_latinoamerica_by_income).mark_circle(
    opacity=0.8,
    stroke='black',
    strokeWidth=1,
    strokeOpacity=0.4
).encode(
            x = alt.X('population', title = 'Population (Millions)', axis = alt.Axis(labelAngle=0)),
            y = alt.Y('income', title = 'AVG Income (Thousand USD/Yr)', scale=alt.Scale(domain=[12, 23])),
            size = alt.Size('income', title = 'AVG Income', scale=alt.Scale(domain=[12, 23])),
            color = alt.Color(field='country', title='Country')
        ).properties(
            width = 400, 
            title = f"Top 10 Highest AVG Income/Yr in Latinoamerica Countries"
        )
```
